## Problem Statement: Recognizing Handwritten Alphabets from Image Data
The goal of this project is to preprocess image files from the MNIST dataset and create a pandas dataframe for classification. Subsequently, a model will be built to accurately recognize handwritten characters.

## Data Preprocessing

```python
import os
import glob
from PIL import Image
import numpy as np
import tqdm

# Define the path to the dataset
dataset_path = r"D:\mnist\New folder"

# Initialize the lists to hold the data and labels
X = []
y = []

# Replace function for thresholding
def replace(p):
    return 255 if p > 50 else 0

# Get all the folder paths
folder_paths = [os.path.join(dataset_path, folder) for folder in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, folder))]

# Debug: Print folder paths
print("Folder paths:", folder_paths)

# Crawling through all the folders
for each_folder in tqdm.tqdm(folder_paths):
    class_name = os.path.basename(each_folder)
    # Fetching all the PNG files from each folder
    png_files = glob.glob(os.path.join(each_folder, "*.png"))

    # Debug: Print number of png files found in each folder
    print(f"Found {len(png_files)} images in folder {each_folder}")

    # Debug: Print the first few image file paths (if any)
    if png_files:
        print("Sample image paths:", png_files[:5])

    # Crawling through all the png files
    for img_path in png_files:
        # Read the image
        img = Image.open(img_path)
        # Apply thresholding to 20% of images
        if np.random.randint(1, 10) <= 4:
            img = img.point(replace)
        # Convert the image into numpy array
        img = np.asarray(img)
        # Flatten the image
        img = img.flatten()
        # Append the image data
        X.append(img)
        # Append the class name
        y.append(class_name)

# Convert lists to numpy arrays
X = np.array(X)
y = np.array(y)

# Print the shapes of the data and labels arrays
print(X.shape)
print(y.shape)
```

## Split X and y

```python
X = dataset.drop("label", axis = 1)
```

```python
y = dataset["label"]
```

## Split into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

## Build the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize models
logreg = LogisticRegression(max_iter=1000)
dtree = DecisionTreeClassifier()
rf = RandomForestClassifier()
svc = SVC()
knn = KNeighborsClassifier()

# Train models
logreg.fit(X_train, y_train)
dtree.fit(X_train, y_train)
rf.fit(X_train, y_train)
svc.fit(X_train, y_train)
knn.fit(X_train, y_train)

# Make predictions
logreg_pred = logreg.predict(X_test)
dtree_pred = dtree.predict(X_test)
rf_pred = rf.predict(X_test)
svc_pred = svc.predict(X_test)
knn_pred = knn.predict(X_test)

# Evaluate models
models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'Support Vector Machine', 'K-Nearest Neighbors']
predictions = [logreg_pred, dtree_pred, rf_pred, svc_pred, knn_pred]

for i, model in enumerate(models):
    print(f"Model: {model}")
    print(f"Accuracy: {accuracy_score(y_test, predictions[i])}")
    print("Classification Report:")
    print(classification_report(y_test, predictions[i]))
    print("Confusion Matrix:")
    cm = confusion_matrix(y_test, predictions[i])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{model} Confusion Matrix")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

#### Support Vector Machine (SVM) achieved the highest accuracy (0.9712) and consistent performance across all metrics.
#### Random Forest also performed very well with slightly lower metrics than SVM.
#### Logistic Regression and Decision Tree had lower accuracy and performance compared to SVM and RF but still performed well.
#### K-Nearest Neighbors had high accuracy and good performance but slightly lower than SVM and RF.

## Real time prediction

```python
from PIL import Image
```

```python
threshold = 128
```

```python
def replace(pixel):
    if pixel > threshold:
        return 0
    else:
        return 255
```

```python

def predict_digit(img_path):
    #read the image
    img = Image.open(img_path)
    #convert into grayscal
    img = img.convert("L")
    img = img.point(replace)
    #resize the image
    img = img.resize((28,28))
    #displays the image
    display(img)
    #convert into numpy array
    img = np.asarray(img)
    #flatten te image to shaep --794
    img = img.flatten()
    #reshape -- (1,784) --1 One image, 784 pixels
    img = img.reshape(1,784)
    #apply scaling
    img = sc.transform(img)
    #feed the imag too the model
    prediction = svc.predict(img)
    return prediction
```

Images were processed by converting them to grayscale, resizing them to 28x28 pixels, and then applying scaling using a predefined scaling method. The prediction was made using a Support Vector Machine (SVM) model.

Upon feeding the image located at 'img_path' into the prediction function, the model correctly predicted the alphabet as 'Y'."

This demonstrates the effectiveness of the SVM model in correctly identifying handwritten characters.
